```python
from pandas import Grouper
train.groupby(Grouper(key='DATE', freq='M')).sum().head()
```

```python
train.groupby(Grouper(key='DATE', freq='M')).sum().plot.line(y=['NUMBER OF PERSONS INJURED', 
                                                                'NUMBER OF PERSONS KILLED'],
                                                            secondary_y=1)
```
